# league_similarity()
### See similar leagues to a focal league
##### league: focal league
##### season: currently 22-23 or 2023
##### nlgs: the number of leagues to show. Default is 20
### This function outputs 3 things: a dataframe with the similar leagues, information that's pertinent to the data, and a figure with the distribution of similar leagues

```python
# This is how to grab all outputs
df, info, dist_fig = league_similarity(
    league = "A-League Men",
    season = "22-23",
    nlgs = 15
)

# This is one way to print the information and then the similarity dataframe
for i in range(len(info)):
    print(info[i])
df
```

# player_similarity()
### See similar players to a focal player, will only show top 2% similarity scores no matter the nplayers
##### player: focal player ("Player_Team" column on available_players() function)
##### position: focal player's position ("Real Position" column on available_players() function)
##### nplayers: the number of teams to show. Default is 20
##### t5_leagues: enter 'y' if you want to only show players in both top 2% of similarity AND currently in UEFA Top 5 leagues
##### similar_lg_team: True/False, to show players only in relatively similar teams & leagues to the focal players' league & team
##### mean_sim: True/False, to sort by average player, league, and team style similarities
### This function outputs 3 things: a dataframe with the similar teams, information that's pertinent to the data, and a figure with the distribution of similar teams

```python
# This is how to grab all outputs
df, info, dist_fig = player_similarity(
    player = "C. Nieuwenhof (22, Western Sydney Wanderers, A-League Men 22-23)",
    position = "CM",
    nplayers = 20,
    t5_leagues = 'n',
    min_age = 1,
    max_age = 99,
    similar_lg_team = False,
    mean_sim = False
)

# This is one way to print the information and then the similarity dataframe
for i in range(len(info)):
    print(info[i])
df
```

```python
dist_fig
```
